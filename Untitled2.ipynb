```python

from gtts import gTTS
import json
import os
from pathlib import Path

# Create output directory
output_dir = '/content/navigation_audio'
os.makedirs(output_dir, exist_ok=True)

print("✓ Setup complete")
```

```python

import json

# Specify your local navigation.json file path
json_path = 'navigasi.json'  # Update this path

# Parse JSON
with open(json_path, 'r', encoding='utf-8') as f:
    navigation_data = json.load(f)

json_filename = json_path.split('/')[-1]
print(f"✓ Loaded: {json_filename}")
print(f"✓ Project: {navigation_data['project']}")
```

```python


def create_audio(text, filename, language='id'):
    """Create MP3 from text"""
    try:
        tts = gTTS(text=text, lang=language, slow=False)
        filepath = os.path.join(output_dir, filename)
        tts.save(filepath)
        return True
    except Exception as e:
        print(f"Error: {e}")
        return False

# Process all locations
total_files = 0
processed_locations = 0

for floor_key, floor_data in navigation_data['floors'].items():
    floor_name = floor_data['floor_name']
    print(f"\n📍 {floor_name}")

    for destination in floor_data['destinations']:
        dest_id = destination['id']
        dest_name = destination['name']
        steps = destination['steps']

        # Combine all steps into one audio file
        intro = f"Navigasi ke {dest_name}. "
        step_text = " ".join([step['text'] for step in steps])
        full_text = intro + step_text

        filename = f"{dest_id}.mp3"

        if create_audio(full_text, filename):
            print(f"  ✓ {dest_name}")
            total_files += 1
            processed_locations += 1
        else:
            print(f"  ✗ {dest_name}")

print(f"\n✅ Conversion complete!")
print(f"Total files created: {total_files}")
```

```python
# List all created files
print(f"\n📂 Generated audio files in {output_dir}:\n")

audio_files = sorted([f for f in os.listdir(output_dir) if f.endswith('.mp3')])
total_size = 0

for i, filename in enumerate(audio_files, 1):
    filepath = os.path.join(output_dir, filename)
    size = os.path.getsize(filepath) / 1024  # KB
    total_size += size
    print(f"{i}. {filename} ({size:.0f} KB)")

print(f"\n{'='*50}")
print(f"✅ Audio generation complete!")
print(f"Total files: {len(audio_files)}")
print(f"Total size: {total_size:.0f} KB")
print(f"Location: {os.path.abspath(output_dir)}")
print(f"{'='*50}")

# Open the output folder (optional - comment out if not needed)
import subprocess
import platform

try:
    if platform.system() == 'Windows':
        os.startfile(os.path.abspath(output_dir))
    elif platform.system() == 'Darwin':  # macOS
        subprocess.Popen(['open', os.path.abspath(output_dir)])
    else:  # Linux
        subprocess.Popen(['xdg-open', os.path.abspath(output_dir)])
    print("\n📁 Opening folder...")
except Exception as e:
    print(f"\nTo access files, open: {os.path.abspath(output_dir)}")
```

```python

# Create a JSON file mapping destinations to audio files
manifest = {
    "project": navigation_data['project'],
    "created_files": {},
    "total_audio_files": len(audio_files)
}

for floor_key, floor_data in navigation_data['floors'].items():
    for destination in floor_data['destinations']:
        dest_id = destination['id']
        dest_name = destination['name']

        manifest["created_files"][dest_id] = {
            "name": dest_name,
            "floor": floor_data['floor_name'],
            "audio_file": f"{dest_id}.mp3",
            "steps": len(destination['steps'])
        }

# Save manifest
manifest_path = os.path.join(output_dir, 'manifest.json')
with open(manifest_path, 'w', encoding='utf-8') as f:
    json.dump(manifest, f, indent=2, ensure_ascii=False)

print("✓ Manifest created: manifest.json")

```
